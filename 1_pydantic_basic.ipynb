```python
from pydantic import BaseModel


class Post(BaseModel):
    id: int
    title: str
    category: str
    is_active: bool
```

```python
post = Post(id=123, title="Hello World", category="Tech")
```

```python
print(post.title)
```

```python
post_dict = {"id": 123, "title": "hello world", "category": "tech"}

post = Post(**post_dict)  # Post(id=123, title="hello world", category="tech")
```

```python
post = Post(id="123", title="hello world", category="1234", is_active="0")
```

```python
print(post.is_active)
```

```python
print(type(post.id))
```
